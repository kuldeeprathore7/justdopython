Q1. The dataframe df has 17 columns. The column total_revenue contains total revenues and column revenue_json contains the breakdown of the total_revenue in multiple component. Write a code to calculcate the percentage contribution of dfp component in total revenue. This should run in less than 1 minute

```python
total_revenue = ['98.46', '42.24', '34.21', '66.32', '72.72']
# revenue_json = {'dfp': [3.49,3.29,2.19,2.69,2.89], 'ga': [3.19,2.29,1.19,2.39,2.19], 'avantis': [1.19,2.59,1.18,2.22,2.17]}
revenue_json = ["{'dfp':3.49, 'ga':3.19, 'avantis':1.19}", "{'dfp':3.29, 'ga':2.29, 'avantis':2.59}", "{'dfp':2.49, 'ga':3.19, 'avantis':2.19}", "{'dfp':2.29, 'ga':4.19, 'avantis':3.19}", "{'dfp':2.59, 'ga':2.14, 'avantis':5.19}"]


```

```python
df = pd.DataFrame({'total_revenue':total_revenue, 'revenue_json':revenue_json})
df
```

```python
df['revenue_json'] = df['revenue_json'].apply(lambda x: eval(x))
df
```

```python
df['dfp_contribution'] = df['revenue_json'].apply(lambda x: x.get('dfp'))
df
```

```python
# print(df.dtypes)
```

```python
df['dfp_contribution_percent'] = 100.0*df['dfp_contribution']/df['total_revenue'].astype(float)
df
```

```python
answer = df['dfp_contribution'].sum()/df['dfp_contribution'].count()
answer

# answer = df['dfp_contribution_percent'].mean()
# answer
```
